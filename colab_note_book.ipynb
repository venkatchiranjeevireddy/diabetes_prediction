Imbalanced Data in Machine Learning
In imbalanced data, the distribution of the target classes is highly skewed. This means that one class has many more instances than the other. This issue is common in many real-world datasets.

For example, in a binary classification problem where the goal is to predict whether a person has a disease or not:

Class 0 (No Disease): 95% of the data
Class 1 (Disease): 5% of the data
This imbalance causes the model to be biased toward the majority class, leading to poor performance, especially on the minority class. In such cases, the model might predict the majority class most of the time, even if it is wrong, and still achieve a high accuracy score, which is misleading.

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x_train_smote,y_train_smote=smote.fit_resample(x_train,y_train)
print(y_train_smote.value_counts())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, RandomizedSearchCV, RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix  # âœ… FIXED HERE

# Your existing code follows...

random_gbc = {
    "n_estimators": [50, 100, 200],
    "learning_rate": [0.01, 0.1, 0.2],
    "max_depth": [3, 4, 5, 6],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 3, 5]
}

random_rf = {
    "n_estimators": [100, 200, 500],
    "max_features": ['auto', 'sqrt', 'log2'],
    "max_depth": list(range(3, 10)),
    "criterion": ['gini', 'entropy']
}

# Model Training
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=2)

# Gradient Boosting
gbc_rs = RandomizedSearchCV(GradientBoostingClassifier(), param_distributions=random_gbc, cv=cv, n_iter=10, n_jobs=-1, verbose=2)
gbc_rs.fit(x_train, y_train)
gbc_best = gbc_rs.best_estimator_
gbc_pred = gbc_best.predict(x_test)
gbc_score = accuracy_score(y_test, gbc_pred)

# Random Forest
rf_rs = RandomizedSearchCV(RandomForestClassifier(), param_distributions=random_rf, cv=cv, n_iter=10, n_jobs=-1, verbose=2)
rf_rs.fit(x_train, y_train)
rf_best = rf_rs.best_estimator_
rf_pred = rf_best.predict(x_test)
rf_score = accuracy_score(y_test, rf_pred)

# Results
print("Gradient Boosting Best Params:", gbc_rs.best_params_)
print("Gradient Boosting Accuracy:", gbc_score)
print("Random Forest Best Params:", rf_rs.best_params_)
print("Random Forest Accuracy:", rf_score)

# Confusion Matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.heatmap(confusion_matrix(y_test, gbc_pred), annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title("Gradient Boosting Confusion Matrix")
sns.heatmap(confusion_matrix(y_test, rf_pred), annot=True, fmt='d', cmap='Greens', ax=axes[1])
axes[1].set_title("Random Forest Confusion Matrix")
plt.show()

# Classification Reports
print("Gradient Boosting Report:\n", classification_report(y_test, gbc_pred))
print("Random Forest Report:\n", classification_report(y_test, rf_pred))
```
